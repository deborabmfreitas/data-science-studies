# Funções lineares

- Funções representadas por uma reta
- Descrevem uma relação direta entre duas variáveis

É representada por:


<div style="text-align: center; font-size: 28px;">
    $f(x) = ax + b$
</div>

onde,

*f(x)* → função de x, ou *y*

*a* → coeficiente angular (inclinação da reta)

*b* → coeficiente linear (intercepto)

# Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Visualizando as funções lineares

```python
def funcao_linear(x, a, b):

    y = a* x + b
    
    return y
```

```python
x = np.linspace(-5, 5, 11)
```

```python
funcao_linear(a = -2, b = 3, x= x)
```

## Diretamente proporcional

O *a* é positivo.

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=1, b=1), marker='o', label='a=-1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Inversamente proporcional

O *a* é negativo.

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=-1, b=1), marker='x', label='a=-1', c= 'darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Comparando as diferentes inclinações

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=-1, b=1), marker='o', label='a=-1')
plt.plot(x, funcao_linear(x, a=1, b=1), marker='x', label='a=1', c='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Comparando as diferentes interceptos

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=1, b=40), marker='o', label='a=-1')
plt.plot(x, funcao_linear(x, a=1, b=20), marker='x', label='a=1', c='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

# Funções quadráticas

- Funções representadas por parábolas
- Descreve relações não-lineares entre duas variáveis

É representada por:


<div style="text-align: center; font-size: 28px;">
    $f(x) = ax² + bx + c$
</div>


onde,

*f(x)* → função de x, ou *y*

*a* → concavidade da parábola

*b* → posição da curva no eixo horizontal (x)

*c* → ponto onde a curva intercepta o eixo vertical (y)

```python
def funcao_quadratica(x, a, b, c):
    
    y = a* x**2 + (b * x) + c
    
    return y
```

```python
x = np.linspace(-10, 10, 51)
```

```python
funcao_quadratica(x= x, a = 9, b = 2, c = -3)
```

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_quadratica(x, a=1, b=1, c= 0), marker='x', label='a=-1')
plt.plot(x, funcao_quadratica(x, a=-1, b=1, c= 0), marker='x', label='a=-1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Quadrática')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```
