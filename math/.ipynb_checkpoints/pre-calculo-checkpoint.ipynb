# Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Função linear

- Funções representadas por uma reta
- Descrevem uma relação direta entre duas variáveis

É representada por:


<div style="text-align: center; font-size: 28px;">
    $f(x) = ax + b$
</div>

onde,

$f(x)$ → função de x, ou *y*

*a* → coeficiente angular (inclinação da reta)

*b* → coeficiente linear (intercepto)

# Visualizando as funções lineares

```python
def funcao_linear(x, a, b):

    y = a* x + b
    
    return y
```

```python
x = np.linspace(-5, 5, 11)
```

```python
funcao_linear(a = -2, b = 3, x= x)
```

## Diretamente proporcional

O *a* é positivo.

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=1, b=1), marker='o', label='a=-1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Inversamente proporcional

O *a* é negativo.

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=-1, b=1), marker='x', label='a=-1', c= 'darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Comparando as diferentes inclinações

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=-1, b=1), marker='o', label='a=-1')
plt.plot(x, funcao_linear(x, a=1, b=1), marker='x', label='a=1', c='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

## Comparando as diferentes interceptos

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_linear(x, a=1, b=40), marker='o', label='a=-1')
plt.plot(x, funcao_linear(x, a=1, b=20), marker='x', label='a=1', c='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Linear')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

# Função quadrática

- Funções representadas por parábolas
- Descreve relações não-lineares entre duas variáveis

É representada por:


<div style="text-align: center; font-size: 28px;">
    $f(x) = ax² + bx + c$
</div>


onde,

$f(x)$  → função de x, ou *y*

*a* → concavidade da parábola

*b* → posição da curva no eixo horizontal (x)

*c* → ponto onde a curva intercepta o eixo vertical (y)

```python
def funcao_quadratica(x, a, b, c):
    
    y = a* x**2 + (b * x) + c
    
    return y
```

```python
x = np.linspace(-10, 10, 51)
```

```python
funcao_quadratica(x= x, a = 9, b = 2, c = -3)
```

```python
plt.figure(figsize= (10,8))
plt.plot(x, funcao_quadratica(x, a=1, b=1, c= 0), marker='x', label='a=-1')
plt.plot(x, funcao_quadratica(x, a=-1, b=1, c= 0), marker='x', label='a=-1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Quadrática')
plt.axhline(c='k')
plt.axvline(c='k')
plt.legend()
plt.grid()
plt.show()
```

# Função polinomial


- Descrevem relações não-lineares entre duas variáveis
- Representadas por expressões matemáticas que envolvem potências de uma variável

É representada por:

<div style="text-align: center; font-size: 28px;">
    $f(x) = a_nx^n + a_{n-1}x^{n-1} + ... + a_1x + a_0$
</div>

onde,  


$f(x)$ → função de x, ou *y*  


$a_nx^n + a_{n-1}x^{n-1} + ... + a_1x + a_0$ → coeficientes que multiplicam as potências correspondentes de x


A potência mais alta de **x**, representada por **n**, determina o grau da função polinomial. 


```python
def calcular_polinomio(coeficientes, valores_x):
    resultados = []

    for x in valores_x:
        resultado = 0
        for i, coeficiente in enumerate(coeficientes):
            resultado += coeficiente * (x ** i)
        resultados.append(resultado)

    return resultados
```

```python
def plotar_grafico_polinomial(coeficientes):
    x = np.linspace(-10, 10, 1000)  # Intervalo amplo de valores de x
    y = calcular_polinomio(coeficientes, x)
    
    plt.figure(figsize= (10,8))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico da função polinomial')
    plt.grid(True)
    plt.show()
```

```python
coeficientes = [3, 2, -5, 1]

plotar_grafico_polinomial(coeficientes)
```

# Função exponencial

- Decrescente e decrescente, que depende do valor da base: *a*
- Se a > 0 e a ≠ 1: crescente
- Se 0 < a < 1: decrescente


É representada por:

<div style="text-align: center; font-size: 28px;">
    $f(x) = a^x$
</div>

onde,

$f(x)$ → função de x, ou *y*   

*a*: coeficiente, que é a base

*x*: coeficiente, que é o expoente.

```python
def plotar_grafico_exponencial(a, titulo):
    x = np.linspace(-5, 5, 100)
    y = a ** x
    
    plt.figure(figsize= (10,8))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(titulo)
    plt.grid(True)
    plt.show()
```

```python
# Gráfico crescente: f(x) = 2^x
plotar_grafico_exponencial(2, 'Gráfico da Função Exponencial Crescente: f(x) = 2^x')
```

```python
# Gráfico decrescente: f(x) = (1/2)^x
plotar_grafico_exponencial(0.5, 'Gráfico da Função Exponencial Decrescente: f(x) = (1/2)')
```

# Função logarítmica

- Comportamento dependente da base do logaritmo
- Se a > 0 e a ≠ 1: crescente
- Se 0 < a < 1: decrescente

É representada por:


<div style="text-align: center; font-size: 28px;">
    $f(x) = \log_a(x)$
</div>
onde,

$f(x)$ → função de x, ou y  
a → base do logaritmo  
x → argumento do logaritmo


```python
def plotar_grafico_logaritmico(base, titulo):
    x = np.linspace(0.1, 10, 100)  # definindo valores de x no intervalo [0.1, 10]
    y = np.log(x) / np.log(base)   # cálculo da função logarítmica

    plt.figure(figsize= (10,8))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(titulo)
    plt.grid(True)
    plt.show()
```

```python
plotar_grafico_logaritmico(2, 'Função crescente')
```

```python
plotar_grafico_logaritmico(0.5, 'Função decrescente')
```

# Função trigonométrica

- Descrevem relações entre ângulos e valores de razões trigonométricas


O comportamento crescente ou decrescente depende do tipo de função trigonométrica e do intervalo considerado
As principais funções trigonométricas são:

 
### Função seno:

<div style="text-align: center; font-size: 28px;">
    $f(x) = \sin(x)$
</div>

descreve a razão entre o comprimento do cateto oposto a um ângulo e a hipotenusa em um triângulo retângulo  



### Função cosseno:

<div style="text-align: center; font-size: 28px;">
    $f(x) = \cos(x)$
</div>

descreve a razão entre o comprimento do cateto adjacente a um ângulo e a hipotenusa em um triângulo retângulo  




### Função tangente:

<div style="text-align: center; font-size: 28px;">
    $f(x) = \tan(x)$
</div>

descreve a razão entre o seno e o cosseno de um ângulo.
As funções trigonométricas são representadas por expressões matemáticas que dependem de um ângulo, denotado por x ou θ.

```python
def calcular_seno(angulo):
    return np.sin(angulo)
```

```python
def calcular_cosseno(angulo):
    return np.cos(angulo)
```

```python
def calcular_tangente(angulo):
    return np.tan(angulo)
```

```python
angulo = np.pi / 4  # ângulo de 45 graus

seno = calcular_seno(angulo)
cosseno = calcular_cosseno(angulo)
tangente = calcular_tangente(angulo)

print("Seno:", seno)
print("Cosseno:", cosseno)
print("Tangente:", tangente)
```
